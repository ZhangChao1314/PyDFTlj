```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use(['science','notebook'])
import sys
sys.path.insert(0, '../src/')
from eos import LJEOS, Calculate_VaporLiquidEquilibria
# Author: Elvis do A. Soares
# Github: @elvissoares
# Date: 2023-02-23
```

```python
eos = LJEOS(sigma=1.0,epsilon=1.0,model='MBWR')

[rhocMBWR,kTcMBWR,rhoMBWR,kTMBWR] = Calculate_VaporLiquidEquilibria(eos)
pMBWR = np.array([eos.p(rhoMBWR[i],kTMBWR[i]) for i in range(rhoMBWR[rhoMBWR<rhocMBWR].size)])
```

```python
rhocMBWR,kTcMBWR
```

```python
eos = LJEOS(sigma=1.0,epsilon=1.0,model='NewMBWR')

[rhoc,kTc,rho,kT] = Calculate_VaporLiquidEquilibria(eos)
plj = np.array([eos.p(rho[i],kT[i]) for i in range(rho[rho<rhoc].size)])
```

```python
rhoc,kTc
```
