```python
def ljpotential(r,eps=1.0,sigma=1.0):
    return 4*eps*((sigma/r)**(12)-(sigma/r)**(6))
```

# Defining the fluid properties

```python
# fluid properties
sigma = 1.0
epsilon = 1.0
L = 10.0*sigma
# Temperature and Density 
kT = 0.71
rhob = 0.84
```

# Defining the functional 


```python
# Test the HS functional 
dft = dft1d(fmtmethod='WBI',ljmethod='MMFA',geometry='Spherical')
dft.Set_Geometry(L=L)
dft.Set_FluidProperties(sigma=sigma,epsilon=epsilon)
dft.Set_Temperature(kT)
dft.Set_BulkDensity(rhob)
Vext = ljpotential(dft.r)
dft.Set_External_Potential(Vext)
```

```python
plt.plot(dft.r,dft.Vext,'-',color='k',label=r'w/corr.')
```

```python
alpha = np.arange(0.2,0.6,0.01)
Nsteps = np.empty_like(alpha)
```

```python
for i in range(alpha.size):
    dft.Set_InitialCondition()
    dft.Calculate_Equilibrium(alpha0=alpha[i],dt=0.02,logoutput=False)
    Nsteps[i] = dft.Niter
```

```python
np.min(Nsteps)
```

```python
alpha[np.argmin(Nsteps)]
```

```python
plt.plot(alpha,Nsteps)
```

```python
dt = np.arange(0.35,0.45,0.01)
Nsteps2 = np.empty_like(dt)
```

```python
for i in range(dt.size):
    dft.Set_InitialCondition()
    dft.Calculate_Equilibrium(alpha0=0.47,dt=dt[i],method='fire',atol=1.0e-5)
    Nsteps2[i] = dft.Niter
```
