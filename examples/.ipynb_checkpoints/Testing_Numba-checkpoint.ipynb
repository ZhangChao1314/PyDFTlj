```python
import matplotlib.pyplot as plt
plt.style.use(['science','notebook'])
import numpy as np
from numba import  njit, prange, vectorize, float64
from numpy import pi, sinc, exp, log, piecewise, linspace, empty_like
```

```python
def phi2func(eta):
    return piecewise(eta,[eta<=1e-3,eta>1e-3],[lambda eta: 1+eta**2/9,lambda eta:  1+(2*eta-eta**2+2*log(1-eta)*(1-eta))/(3*eta)])
```

```python
@njit
def phi2funcjitted(eta):
    y = empty_like(eta)
    mask = eta <= 1e-3
    y[mask] = 1+eta[mask]**2/9
    mask = eta > 1e-3
    y[mask] = 1+(2*eta[mask]-eta[mask]**2+2*log(1-eta[mask])*(1-eta[mask]))/(3*eta[mask])
    return y
```

```python
@vectorize
def phi2funcjitted2(eta):
    if eta >= 1e-3:
        return 1+(2*eta-eta**2+2*log(1-eta)*(1-eta))/(3*eta)
    else:
        return 1+eta**2/9
```

```python
eta = np.arange(0.0,0.99,1e-4)
```

```python
%timeit -n 10000 phi2func(eta)
```

```python
%timeit -n 10000 phi2funcjitted(eta)
```

```python
%timeit -n 10000 phi2funcjitted2(eta)
```

```python
phi2funcjitted2(eta)==phi2func(eta)
```

```python
# plt.plot(eta,phi2funcjitted(eta))
plt.plot(eta,phi2funcjitted2(eta),'--')
```

```python
@vectorize
def phi3func(eta):
    if eta<=1e-3:
        return 1-2*eta/9-eta**2/18
    else: 
        return 2*(eta+log(1-eta)*(1-eta)**2)/(3*eta**2)        
```

```python
1-4*eta/9 + eta**2/18
```

```python
for i in range(eta.size):
    print(eta[i],phi3func(eta[i]))
```

```python
plt.plot(eta,phi3func(eta),'--')
```
